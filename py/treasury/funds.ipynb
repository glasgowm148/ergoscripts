```python
import matplotlib.pyplot as plt
import numpy as np

# Data to plot
labels = ['SPF', 'RSN', 'Ergo', 'USD Coin', 'Bitcoin']

# List of donations for each currency
donations_SPF = [10000, 5000, 30000]  # Replace these values with your actual data
donations_RSN = [20000, 5000, 10000]
donations_Ergo = [15000, 10000, 5000]
donations_USDC = [10000, 5000, 5000]
donations_Bitcoin = [5000, 10000, 15000]

# Prices for each currency in USD
price_SPF = 0.061569  # Replace these values with your actual data
price_RSN = 0.005005
price_Ergo = 1.47
price_USDC = 1.01  # USDC is a stablecoin, so it's typically around $1
price_Bitcoin = 26416.9

# Calculate total donations for each currency
total_SPF = np.sum(donations_SPF) * price_SPF
total_RSN = np.sum(donations_RSN) * price_RSN
total_Ergo = np.sum(donations_Ergo) * price_Ergo
total_USDC = np.sum(donations_USDC) * price_USDC
total_Bitcoin = np.sum(donations_Bitcoin) * price_Bitcoin

# Prepare data for the pie chart
totals = [total_SPF, total_RSN, total_Ergo, total_USDC, total_Bitcoin]

# Define autopct format function
def autopct_format(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct * total / 100.0))
        return '${:,.2f}'.format(val)
    return my_autopct

# Plot
fig, ax = plt.subplots(figsize=(10, 6))
wedges, texts, autotexts = ax.pie(totals, labels=labels, autopct=autopct_format(totals), startangle=140, textprops={'fontsize': 14})
plt.setp(autotexts, size=12, weight='bold')

# Display total sum in USD
total_sum = sum(totals)
plt.text(0, 0, 'Total:\n${:,.2f}'.format(total_sum), ha='center', va='center', size=12, color='black')

plt.title('Treasury Allocation in USD', fontsize=20)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
