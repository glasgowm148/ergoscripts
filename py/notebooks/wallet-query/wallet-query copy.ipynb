# Pull

API Request to pull the data and then place it into a pandas dataframe. You can analyse the json directly by uncommenting these lines below.

```python
# API Request
# Docs : https://api.ergoplatform.com/api/v1/docs/

import requests
for p in range(1,5):
    url = 'https://api.ergoplatform.com/api/v1/boxes/byAddress/'+target+'?page='+str(p)
    r = requests.get(url)
    
    json = r.json()


```

```python
# load json into a dataframe
df = pd.DataFrame(json['items'])


# for investigating parsing errors 
df.to_csv("EF.csv", index=False)

df.info()
```

# Dataframe

Pandas allows us to view the information easier, we drop the columns we don't need to here to clear up clutter.

```python
# Drop unneeded columns
df.drop(columns=['mainChain','ergoTree','transactionId','boxId', 
                 'blockId','address','index',
                 'additionalRegisters','spentTransactionId'], inplace=True)

```

# Tidy

The UTXO model tracks the total balance on each transaction in nanoErgs. Some small fixes to make it more readable.

```python
# Finding difference
df['BalanceChange (ERG)'] = df.value.diff()

# nanoErg -> Erg
df['value'] = df['value'] / 1000000000
df['BalanceChange (ERG)'] = df['BalanceChange (ERG)'] / 1000000000

# Reversing polarity
df['BalanceChange (ERG)'] = df['BalanceChange (ERG)'] * -1 
df['BalanceChange (ERG)'] = df['BalanceChange (ERG)'].shift(-1)
```
